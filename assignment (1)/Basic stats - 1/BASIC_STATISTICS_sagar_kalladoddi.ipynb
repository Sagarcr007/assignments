#Z-score normalization, also known as standardization, is a technique used to transform numerical data in a dataset so that it follows a standard normal  distribution with a mean of 0 and a standard deviation of 1

```python
sales2=sales.copy()
sales2['Volume']=(sales['Volume']-sales['Volume'].mean())/sales['Volume'].std()
sales2['Avg Price']=(sales['Avg Price']-sales['Avg Price'].mean())/sales['Avg Price'].std()
sales2['Total Sales Value']=(sales['Total Sales Value']-sales['Total Sales Value'].mean())/sales['Total Sales Value'].std()
sales2['Discount Rate (%)']=(sales['Discount Rate (%)']-sales['Discount Rate (%)'].mean())/sales['Discount Rate (%)'].std()
sales2['Discount Amount']=(sales['Discount Amount']-sales['Discount Amount'].mean())/sales['Discount Amount'].std()
sales2['Net Sales Value']=(sales['Net Sales Value']-sales['Net Sales Value'].mean())/sales['Net Sales Value'].std()
sales2.head()
```

Conversion of Categorical Data into Dummy Variables

```python
sales3=sales2.copy()
sales3=pd.get_dummies(sales3,columns=['Day','City','BU','Brand','Model'])
sales3.head()
```
