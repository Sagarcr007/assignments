K-Means


```python
from sklearn.cluster import KMeans
```

```python
def norm_func(i):
  x=(i-i.mean())/i.std()
  return(x)
```

```python
x=norm_func(data.iloc[:,1:])
x
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
print(wcss)
plt.plot(range(1,11),wcss)
plt.title('elbow method')
plt.xlabel('no of clusters')
plt.ylabel('wcss(Within Cluster Sum of Squares)')
plt.show()
```

```python
cluster_new=KMeans(6,random_state=42)
cluster_new.fit(x)
```

```python
cluster_new.labels_
```

```python
data['clusterid_new']=cluster_new.labels_
data
```

```python
cluster_new.cluster_centers_
```

```python
data.groupby('clusterid_new').agg(['mean']).reset_index() #cluster 0
```

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette score for K-Means
silhouette_kmeans = silhouette_score(data, kmeans.labels_)
print("Silhouette Score for K-Means:", silhouette_kmeans)
```

```python
cluster_counts=data['clusterid_new'].value_counts().sort_index()

color=['r','g','b','y','m','c']
plt.scatter(cluster_counts.index, cluster_counts.values,color=color)
plt.xlabel('Cluster_new')
plt.ylabel('number of points')
plt.title('Cluster distribution')
plt.show()
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
array=data.values
array
```

```python
stscaler=StandardScaler().fit(array)
x=stscaler.transform(array)
x
```

```python
stscaler = StandardScaler()
stscaler.fit(array)  # Fit the scaler to your data
x = stscaler.transform(array)  # Now transform the data
x
```

```python
db=DBSCAN(eps=3,min_samples=54)
db.fit(x)  # Fit the model to your data
```

```python
db.labels_
```

```python
db=pd.DataFrame(db.labels_,columns=['cluster'])
db  #cluster 0
```

```python
data['cluster']=pd.DataFrame(db.cluster)
data
```

hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# normalization function
def minmax_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=minmax_func(data.iloc[:,1:])
df_norm
```

```python
df_norm
```

```python
# single  ,compelet, average,centroid
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='centroid'))
```

```python
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
data['h_clusterid']=y_hc
data
```
